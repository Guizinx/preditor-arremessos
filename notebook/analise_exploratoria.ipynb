```python
# Criando boxplots
plt.figure(figsize=(15, 8))
for i, col in enumerate(colunas, 1):
    plt.subplot(3, 3, i)
    sns.boxplot(data=df, y=col, color='skyblue')
    plt.title(f'Boxplot - {col}')

plt.tight_layout()
plt.show()
```
